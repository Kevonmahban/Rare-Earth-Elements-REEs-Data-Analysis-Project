```python
deposit_type_counts = df['Dep_Type'].value_counts()
print("Most common Deposit Types:")
print(deposit_type_counts)
```

In this code, I use df['Dep_Type'].value_counts() to count the occurrences of each unique value in the 'Dep_Type' column. This helps identify the most common deposit types in the dataset, and the result is printed to show the frequency of each deposit type.



```python
plt.figure(figsize=(9,8))
sns.countplot(data=df, y='Dep_Type', order=deposit_type_counts.index)
plt.title('Most Common REE Deposit Types')
plt.xlabel('Count')
plt.ylabel('Deposit Type')
plt.tight_layout()
plt.show()
```

In this code, I create a count plot using Seaborn to visualize the distribution of different REE deposit types. The sns.countplot() function plots the count of each deposit type on the x-axis, ordered by their frequency from deposit_type_counts.index.
```python
country_counts = df['Country'].value_counts()

plt.figure(figsize=(8,6))
sns.barplot(x=country_counts.values[:10], y=country_counts.index[:10], palette='mako')
plt.title('Top 10 Countries with REE Occurrences')
plt.xlabel('Number of Occurrences')
plt.ylabel('Country')
plt.show()
```
