```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(
    x=['Monazite','Euxenite'],
    y=[
            df[df['Material'].str.contains('monazite')]['CntSIComAm'].sum(),
            df[df['Material'].str.contains('euxenite')]['CntSIComAm'].sum()
])
plt.title('Total Contained Commodity by Mineral')
plt.xlabel('Mineral')
plt.ylabel('Contained Amt In Metric Tons')
```

```python
sns.barplot(
    x=['Monazite','Euxenite','Iron'],y=[
            df[df['Material'].str.contains('monazite')]['Year_To'].max(),
            df[df['Material'].str.contains('euxenite')]['Year_To'].max(),
            df[df['Material'].str.contains('iron')]['Year_To'].max()
            ], color = 'Grey'
)
plt.ylim(1900,2025)
plt.ylabel('Year')
plt.xlabel('Mineral')
plt.title('Last Year of Recorded Mining by Minerals')
```
